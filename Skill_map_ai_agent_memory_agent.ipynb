```python
import requests
import json
from pprint import pprint
from langchain.tools import tool
from langchain.agents import create_agent
from google.colab import userdata

# -----------------------------
# TOOL 1 : Skill Demand Tool
# -----------------------------
@tool
def skill_demand_tool(skill: str) -> str:
    """Check demand for a skill in the industry."""
    print("\nChecking industry demand for:", skill)

    demand_data = {
        "generative ai": "Very High demand. Companies are hiring for LLM, AI Engineers, and Prompt Engineers.",
        "python": "High demand. Widely used for backend, AI, and data science.",
        "machine learning": "High demand in AI, analytics, and product companies."
    }

    return demand_data.get(skill.lower(), "Demand data not available")


# -----------------------------
# TOOL 2 : Job Search Tool
# -----------------------------
@tool
def search_jobs(skill: str, location: str) -> list:
    """Search for jobs requiring a specific skill."""

    print("\nCalling search_jobs tool")
    print(f"Searching jobs for: {skill} in {location}")

    rapidapi_key = userdata.get("RAPIDAPI_KEY").strip()

    url = "https://jsearch.p.rapidapi.com/search"

    headers = {
        "x-rapidapi-key": rapidapi_key,
        "x-rapidapi-host": "jsearch.p.rapidapi.com"
    }

    querystring = {
        "query": f"{skill} in {location}",
        "page": "1",
        "country": "in",
        "employment_types": "INTERN,FULLTIME",
        "job_requirements": "no_experience,under_3_years_experience"
    }

    response = requests.get(url, headers=headers, params=querystring)
    data = response.json()

    jobs = data.get("data", [])

    result = []
    for job in jobs[:5]:   # return top 5 jobs
        result.append({
            "title": job.get("job_title"),
            "company": job.get("employer_name"),
            "location": job.get("job_city"),
            "apply_link": job.get("job_apply_link")
        })

    print(f"Found {len(result)} jobs\n")
    return result


# -----------------------------
# SYSTEM PROMPT
# -----------------------------
system_prompt = """
You are a helpful AI job assistant.

Steps:
1. Use skill_demand_tool to check industry demand.
2. Use search_jobs to find jobs.
3. Always return job openings as a Python list of dictionaries.

Example format:
[
 {"title": "...", "company": "...", "location": "...", "apply_link": "..."}
]
"""


# -----------------------------
# CREATE AGENT
# -----------------------------
agent = create_agent(
    model=model,
    tools=[skill_demand_tool, search_jobs],
    system_prompt=system_prompt,
    debug=True
)


# -----------------------------
# USER QUERY
# -----------------------------
user_query = "What's the demand for generative ai in the industry and show me related job openings in India"


# -----------------------------
# RUN AGENT
# -----------------------------
response = agent.invoke({
    "messages": [{"role": "user", "content": user_query}]
})


# -----------------------------
# PRINT RESPONSE
# -----------------------------


final_msg = response["messages"][-1].content

# extract text
text_output = final_msg[0]["text"]

# split demand and jobs
parts = text_output.split("[")

demand = parts[0]
jobs_json = "[" + parts[1]

jobs = json.loads(jobs_json)

print("\nDemand Info:")
print(demand)

print("\nJob Openings:\n")

for job in jobs:
    print(f"Title   : {job['title']}")
    print(f"Company : {job['company']}")
    print(f"Location: {job['location']}")
    print(f"Apply   : {job['apply_link']}")
    print("-"*50)
```

```python
import requests
from google.colab import userdata
from langchain.tools import tool
from langchain_tavily import TavilySearch
from langchain.chat_models import init_chat_model
from langchain.agents import create_agent
from langgraph.checkpoint.memory import InMemorySaver

TAVILY_API_KEY = (userdata.get("TAVILY_API_KEY") or "").strip()
RAPIDAPI_KEY = (userdata.get("RAPIDAPI_KEY") or "").strip()
GROQ_API_KEY = (userdata.get("GROQ_API_KEY") or "").strip()

if not TAVILY_API_KEY:
    raise ValueError("Missing TAVILY_API_KEY")
if not RAPIDAPI_KEY:
    raise ValueError("Missing RAPIDAPI_KEY")
if not GROQ_API_KEY:
    raise ValueError("Missing GROQ_API_KEY")

skill_demand_tool = TavilySearch(
    max_results=5,
    search_depth="advanced",
    tavily_api_key=TAVILY_API_KEY,
)

@tool
def search_jobs(skill: str, location: str) -> list:
    """Search for jobs based on skill and location"""

    url = "https://jsearch.p.rapidapi.com/search"

    headers = {
        "x-rapidapi-key": str(RAPIDAPI_KEY),
        "x-rapidapi-host": "jsearch.p.rapidapi.com",
    }

    params = {
        "query": f"{skill} in {location}",
        "page": "1",
        "num_pages": "1",
        "country": "in",
        "employment_types": "INTERN,FULLTIME",
        "job_requirements": "no_experience,under_3_years_experience",
    }

    response = requests.get(url, headers=headers, params=params)
    data = response.json()

    jobs = data.get("data", [])

    return [
        {
            "title": job.get("job_title"),
            "company": job.get("employer_name"),
            "location": job.get("job_city"),
            "apply_link": job.get("job_apply_link"),
        }
        for job in jobs
    ]

SYSTEM_PROMPT = """
You are a Skill-to-Career Mapping assistant that helps students understand skill demand
and find matching job opportunities.

Use tools when needed.

Give output in clean readable format with sections and spacing.
Include job title, company, location, and apply link clearly.
Do not use markdown formatting.
"""

model = init_chat_model(
    "groq:llama-3.3-70b-versatile",
    api_key=GROQ_API_KEY,
)

checkpointer = InMemorySaver()
config = {"configurable": {"thread_id": "1"}}

agent = create_agent(
    model=model,
    system_prompt=SYSTEM_PROMPT,
    tools=[skill_demand_tool, search_jobs],
    checkpointer=checkpointer,
)

user_query = "What's the demand for generative AI and show jobs in India"

response = agent.invoke(
    {"messages": [{"role": "user", "content": user_query}]},
    config=config,
)

print(response["messages"][-1].content)

user_query = "Tell me more about the second job"

response = agent.invoke(
    {"messages": [{"role": "user", "content": user_query}]},
    config=config,
)

print(response["messages"][-1].content)
```
